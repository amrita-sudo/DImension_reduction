```python
import numpy as np
import pandas as pd
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA

```

```python
from sklearn.datasets import load_wine
data_df=load_wine()
data_df
```

```python

x=data_df.data
y=data_df.target
print("Feature array shape:",x.shape)
print("Target array shape:",y.shape)
```

```python
from sklearn.preprocessing import StandardScaler
x_std=StandardScaler().fit_transform(x)
x_std
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_std,y,test_size=0.3,random_state=0)
```

```python
classifier=RandomForestClassifier(max_depth=2,random_state=0)
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
```

```python
cm=confusion_matrix(y_test,y_pred)
print("Confusion_matrix:\n",cm)
print("Accuracy:\n",accuracy_score(y_test,y_pred))

```

```python
#Create LDA instance and transform x_train and x_test
lda=LDA(n_components=2)
lda_x_train=lda.fit_transform(x_train,y_train)
lda_x_test=lda.transform(x_test)
```

```python
new_classifier=RandomForestClassifier(max_depth=2,random_state=0)
new_classifier.fit(lda_x_train,y_train)
new_y_pred=new_classifier.predict(lda_x_test)
```

```python
new_cm=confusion_matrix(y_test,new_y_pred)
print("Confusion_matrix:\n",new_cm)
print("Accuracy:\n",accuracy_score(y_test,new_y_pred))
```
